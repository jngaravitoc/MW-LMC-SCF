```python
rlmc1, powerlmc1, vzlmc1 = np.loadtxt('../data/interim/r_dipole_vals_lmc1.txt', skiprows=1, unpack=True)
rlmc2, powerlmc2, vzlmc2 = np.loadtxt('../data/interim/r_dipole_vals_lmc2.txt', skiprows=1, unpack=True)
rlmc3, powerlmc3, vzlmc3 = np.loadtxt('../data/interim/r_dipole_vals_lmc3.txt', skiprows=1, unpack=True)
rlmc4, powerlmc4, vzlmc4 = np.loadtxt('../data/interim/r_dipole_vals_lmc4.txt', skiprows=1, unpack=True)
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(rlmc1, powerlmc1**0.5,  lw=2, label='Sim 5')
plt.plot(rlmc2, powerlmc2**0.5,  lw=2, label='Sim 6')
plt.plot(rlmc3, powerlmc3**0.5,  lw=2, label='Sim 7')
plt.plot(rlmc4, powerlmc4**0.5,  lw=2, label='Sim 8')


legend(fontsize=14)
xlabel(r'$\rm{r}[kpc]$')
ylabel(r'$\sqrt{3\ C_l} \rm{[km/s]}$')
axvline(30, ls=':', c='k')
xticks(np.arange(0, 200, 25))
title('Radial velocity the dipole')
xlim(0, 180)
ylim(0, 150)

savefig('vr_dipole.pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(rlmc1, vzlmc1,  lw=2, label='Sim 5')
plt.plot(rlmc2, vzlmc2,  lw=2, label='Sim 6')
plt.plot(rlmc3, vzlmc3,  lw=2, label='Sim 7')
plt.plot(rlmc4, vzlmc4,  lw=2, label='Sim 8')


legend(fontsize=14)
xlabel(r'$\rm{r}[kpc]$')
ylabel(r'$v_z \rm{[km/s]}$')
axvline(30, ls=':', c='k')
xticks(np.arange(0, 200, 25))
title('Vertical velocity the dipole')
xlim(0, 180)
savefig('vz_dipole.pdf', bbox_inches='tight')
```
