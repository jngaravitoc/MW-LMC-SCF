```python
r, power = np.loadtxt('../data/interim/r_dipole_vals.txt', skiprows=1, unpack=True)
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(r, power, c='k', lw=2)
xlabel(r'$\rm{r}[kpc]$')
ylabel(r'$3\ C_l \rm{[km^2/s^2]}$')
axvline(30, ls=':', c='k')
xticks(np.arange(0, 200, 25))
title('Power of the dipole in $<v_r>$')
savefig('vr_dipole.pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(r, power**0.5, c='k', lw=2)
xlabel(r'$\rm{r}[kpc]$')
ylabel(r'$v_{dipole}\rm{[km/s]}$')
axvline(30, ls=':', c='k')
xticks(np.arange(0, 200, 25))
yticks(np.arange(0, 120, 20))

#title('Power of the dipole in $<v_r>$')
savefig('vr_vel.pdf', bbox_inches='tight')
```
