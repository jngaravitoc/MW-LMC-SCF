```python
def plot_density(figname, rho, title):
    fig, ax = plt.subplots(1, 1, figsize=(5, 4))
    im = ax.contourf(y_grid, z_grid, np.log10(np.abs(rho)).reshape(bins, bins), levels=np.linspace(-5.5, -1.2, 30),
             origin='lower', cmap='Spectral_r')
    #ax.set_facecolor('k')

    ax.set_xlim(-100, 100)
    ax.set_ylim(-100, 100)
    ax.set_xlabel('y[kpc]')
    ax.set_ylabel('z[kpc]')
    plt.title(title)
    fig.colorbar(im)

    savefig(figname, bbox_inches='tight', dpi=500)
```
