```python
%pylab inline
import biff
import pygadgetreader
```

```python
font = {'size':18, 'family':'serif'}
matplotlib.rc('font', **font)
```

```python
halo_pos = pygadgetreader.readsnap('../data/MWLMC5_100M_new_b1_110', 'pos', 'dm')
```

```python
rand_1E6 = np.random.randint(0, 115000000, 1000000)
rand_1E5 = np.random.randint(0, 115000000, 100000)
rand_1E4 = np.random.randint(0, 115000000, 10000)
rand_1E3 = np.random.randint(0, 115000000, 1000)
rand_1E3_2 = np.random.randint(0, 115000000, 1000)
rand_1E3_3 = np.random.randint(0, 115000000, 1000)
```

```python
halo_pos_1E6 = halo_pos[rand_1E6]
halo_pos_1E5 = halo_pos[rand_1E5]
halo_pos_1E4 = halo_pos[rand_1E4]
halo_pos_1E3 = halo_pos[rand_1E3]
halo_pos_1E3_2 = halo_pos[rand_1E3_2]
halo_pos_1E3_3 = halo_pos[rand_1E3_3]
```

```python
S_1e4, T_1e4 = biff.compute_coeffs_discrete(np.ascontiguousarray(halo_pos_1E4.astype(float)),
                                            nmax=20, lmax=20, mass=np.ones(10000)*10, r_s=40.85,
                                           skip_m=True)
```

```python
S_1e5, T_1e5 = biff.compute_coeffs_discrete(np.ascontiguousarray(halo_pos_1E5.astype(float)),
                                            nmax=20, lmax=20, mass=np.ones(100000), r_s=40.85,
                                           skip_m=True)
```

```python
S_1e3, T_1e3 = biff.compute_coeffs_discrete(np.ascontiguousarray(halo_pos_1E3.astype(float)),
                                            nmax=20, lmax=20, mass=np.ones(1000)*100, r_s=40.85,
                                           skip_m=True)
```

```python
S_1e3_2, T_1e3_2 = biff.compute_coeffs_discrete(np.ascontiguousarray(halo_pos_1E3_2.astype(float)),
                                            nmax=20, lmax=20, mass=np.ones(1000)*100, r_s=40.85,
                                           skip_m=True)

S_1e3_3, T_1e3_3 = biff.compute_coeffs_discrete(np.ascontiguousarray(halo_pos_1E3_3.astype(float)),
                                            nmax=20, lmax=20, mass=np.ones(1000)*100, r_s=40.85,
                                           skip_m=True)
```

```python
figure(figsize=(12, 4.5))
subplot(1, 2, 1)
title('Bias')
plot(np.log10(np.abs(S_1e3[0,:,0])), c='purple', lw=1.5, label='$10^3$ particles')
plot(np.log10(np.abs(S_1e4[0,:,0])), c='purple', ls='--', label='$10^4$ particles')
plot(np.log10(np.abs(S_1e5[0,:,0])), c='purple', ls=':', label='$10^5$ particles')
ylabel(r'$\rm{Log_{10}}|S_{nlm}|$')
xlabel('$l$')
legend(fontsize=16)

subplot(1, 2, 2)
title('Variance')
plot(np.log10(np.abs(S_1e3[0,:,0])), c='purple', lw=1.5, label='Monte-Carlo sample 1')
plot(np.log10(np.abs(S_1e3_2[0,:,0])), c='purple', ls='--', label='Monte-Carlo sample 2')
plot(np.log10(np.abs(S_1e3_3[0,:,0])), c='purple', ls=':', label='Monte-Carlo sample 3')
legend(fontsize=16)
ylabel(r'$\rm{Log_{10}}|S_{nlm}|$')
xlabel('$l$')

savefig('noise_effects_on_BFE.pdf', bbox_inches='tight')
```
